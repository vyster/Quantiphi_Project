```python
def get_dict_to_csv_shot(d)    :
    f = {}
    shot_number_video =-1

    for minute in d['VIDEO']['SCENE']:
        shot_number_minute=-1

        for shot in minute['CAMERA_CUT']:
            shot_number_minute = shot_number_minute + 1
            shot_number_video = shot_number_video + 1
            for frame in shot['FRAME']:

    #             f['STARTTIME'] = (frame['STARTTIME'])
                f[(frame['STARTTIME'])] = {'shot_number_video':shot_number_video,'shot_number_minute':shot_number_minute}
        #             print(frame['STARTTIME'])
        
    print('shot - Done.')
        
    return pd.DataFrame(f).T
```

```python
def get_dict_to_csv(d,feature_name): #feature_name : 'SHOT_ANGLE','LOCATION',SPEECH
    f = {}
    for minute in d['VIDEO']['SCENE']:
        for shot in minute['CAMERA_CUT']:
            for frame in shot['FRAME']:

                f[(frame['STARTTIME'])] = (frame[feature_name])['PREDICTION'][0]
    
    print(feature_name+' - Done.')
        
    return (pd.DataFrame(f).T).rename(columns = { 'CONFIDENCE' : feature_name+"_CONFIDENCE",'LABEL' : feature_name+"_LABEL"})
    
```

```python
 def get_character_from_json2(d):
    f={}
    minute = -1
    for scene in d['VIDEO']['SCENE']:
        minute+=1
        shot = -1
        for camera_cut in scene['CAMERA_CUT']:
            shot+=1
            for frame in camera_cut['FRAME']:
                f.setdefault(frame['STARTTIME'])
                if type(f[(frame['STARTTIME'])]) != dict:
                    f[(frame['STARTTIME'])] = {}
            #     f[frame['STARTTIME']].setdefault('PREDICTION')
                char_number = 0
                for char in frame['AGE_GENDER']:
                    f[frame['STARTTIME']].setdefault(char_number)
                    if type(f[frame['STARTTIME']][char_number]) != dict:
                        
                        f[frame['STARTTIME']][char_number] = {}
                        f[frame['STARTTIME']][char_number]['AGE_GENDER'] = {}
                        f[frame['STARTTIME']][char_number]['CHARACTER'] = {}
                        
                        
                    f[frame['STARTTIME']][char_number]['AGE_GENDER']['PREDICTION'] = char['PREDICTION'][0]
                    f[frame['STARTTIME']][char_number]['YMIN'] = char['YMIN']
                    f[frame['STARTTIME']][char_number]['YMAX'] = char['YMAX']
                    f[frame['STARTTIME']][char_number]['XMAX'] = char['XMAX']
                    f[frame['STARTTIME']][char_number]['XMIN'] = char['XMIN']
                    f[frame['STARTTIME']][char_number]['DLIB_CONFIDENCE'] = char['DLIB_CONFIDENCE']
                    f[frame['STARTTIME']][char_number][' QUALITY_PARAMETER']= char['QUALITY_PARAMETER']

                    f[frame['STARTTIME']][char_number]['MINUTE'] = minute
                    f[frame['STARTTIME']][char_number]['SHOT_ID'] = shot


                    f[frame['STARTTIME']][char_number]['SCENE_STARTTIME'] = scene['STARTTIME']
                    f[frame['STARTTIME']][char_number]['SCENE_ENDTIME'] = scene['ENDTIME']

                    f[frame['STARTTIME']][char_number]['VIDEO_STARTTIME'] = d['VIDEO']['STARTTIME']
                    f[frame['STARTTIME']][char_number]['VIDEO_ENDTIME'] = d['VIDEO']['ENDTIME']
                    f[frame['STARTTIME']][char_number]['VIDEO_ID'] = d['VIDEO']['NAME']
                    
                    
                    
                    
                    char_number +=1
                    
                    
                char_number = 0    
                for char in frame['CHARACTER']:
                    f[frame['STARTTIME']].setdefault(char_number)
                    if type(f[frame['STARTTIME']][char_number]) != dict:
                        
                        f[frame['STARTTIME']][char_number] = {}
                        f[frame['STARTTIME']][char_number]['CHARACTER'] = {}
                        
                        
                    f[frame['STARTTIME']][char_number]['CHARACTER']['PREDICTION'] = char['PREDICTION'][0]
                    f[frame['STARTTIME']]['COLOR_PALETTE'] = ((frame['COLOR'])['COLOR_PALETTE'][0])
                    f[frame['STARTTIME']]['COLOR_PALETTE'] = ((frame['COLOR'])['COLOR_PALETTE'][0])
                    f[frame['STARTTIME']]['COLOR_LUMINANCE'] = ((frame['COLOR'])['COLOR_LUMINANCE'][0])
                    f[frame['STARTTIME']]['COLOR_TEMPERATURE'] = ((frame['COLOR'])['COLOR_TEMPERATURE'][0])

                    f[frame['STARTTIME']]['LOCATION'] = ((frame['LOCATION'])['PREDICTION'][0])
                    f[frame['STARTTIME']]['SHOT_ANGLE'] = ((frame['SHOT_ANGLE'])['PREDICTION'][0])

                    f[frame['STARTTIME']]['SPEECH'] = ((frame['SPEECH'])['PREDICTION'][0])

                    char_number +=1
    
    
    print('get_Char_from_json - Done.')

                    
    return pd.DataFrame(f).T
```

```python
def dict_to_columns(old_df,char_number):
    new_df = old_df.copy()
    char = str(char_number)
    new_df['CHAR_'+char+'_YMIN'] = -1
    new_df['CHAR_'+char+'_YMAX'] = -1
    new_df['CHAR_'+char+'_XMIN'] = -1
    new_df['CHAR_'+char+'_XMAX'] = -1
    new_df['CHAR_'+char+'_CHARACTER_PREDICTION_CONFIDENCE'] = -1
    new_df['CHAR_'+char+'_CHARACTER_PREDICTION_LABEL'] = -1
    new_df['CHAR_'+char+'_DLIB_CONFIDENCE'] = -1
    new_df['CHAR_'+char+'_VIDEO_ID'] = -1
    new_df['CHAR_'+char+'_MINUTE'] = -1
    new_df['CHAR_'+char+'_SHOT_ID'] = -1
    new_df['CHAR_'+char+'_SCENE_STARTTIME'] = -1
    new_df['CHAR_'+char+'_SCENE_ENDTIME'] = -1
    new_df['CHAR_'+char+'_VIDEO_STARTTIME'] = -1
    new_df['CHAR_'+char+'_VIDEO_ENDTIME'] = -1
    new_df['CHAR_'+char+'_QUALITY_PARAMETER_WIDTH'] = -1
    new_df['CHAR_'+char+'_QUALITY_PARAMETER_HEIGHT'] = -1
    new_df['CHAR_'+char+'_QUALITY_PARAMETER_PIXEL_COUNT'] = -1
    new_df['CHAR_'+char+'_AGE_GENDER_PREDICTION_CONFIDENCE'] = -1
    new_df['CHAR_'+char+'_AGE_GENDER_PREDICTION_LABEL'] = -1



    for i in range(len(new_df)):
        item = new_df.iloc[i,new_df.columns.get_loc(char_number)]
        if type(item) == dict:
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_YMIN')] = item.get('YMIN','unexplained')
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_YMAX')] = item.get('YMAX','unexplained')
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_XMIN')] = item.get('XMIN','unexplained')
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_XMAX')] = item.get('XMAX','unexplained')
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_CHARACTER_PREDICTION_CONFIDENCE')] = item['CHARACTER']['PREDICTION']['CONFIDENCE']  if (bool(item['CHARACTER'])) else 'unexplained'
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_CHARACTER_PREDICTION_LABEL')] = item['CHARACTER']['PREDICTION']['LABEL']  if (bool(item['CHARACTER'])) else 'unexplained'
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_DLIB_CONFIDENCE')] = item.get('DLIB_CONFIDENCE','unexplained')
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_VIDEO_ID')] = item.get('VIDEO_ID','unexplained')
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_MINUTE')] = item.get('MINUTE','unexplained')
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_SHOT_ID')] = item.get('SHOT_ID','unexplained')
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_SCENE_STARTTIME')] = item.get('SCENE_STARTTIME','unexplained')
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_SCENE_ENDTIME')] = item.get('SCENE_ENDTIME','unexplained')
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_VIDEO_STARTTIME')] = item.get('VIDEO_STARTTIME','unexplained')
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_VIDEO_ENDTIME')] = item.get('VIDEO_ENDTIME','unexplained')
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_QUALITY_PARAMETER_WIDTH')] = item[' QUALITY_PARAMETER']['WIDTH'] if item.get(' QUALITY_PARAMETER',False) else 'unexplained' 
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_QUALITY_PARAMETER_HEIGHT')] = item[' QUALITY_PARAMETER']['HEIGHT'] if item.get(' QUALITY_PARAMETER',False) else 'unexplained'
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_QUALITY_PARAMETER_PIXEL_COUNT')] = (item[' QUALITY_PARAMETER']['PIXEL_COUNT']) if item.get(' QUALITY_PARAMETER',False) else 'unexplained'
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_AGE_GENDER_PREDICTION_CONFIDENCE')] = (item['AGE_GENDER']['PREDICTION']['CONFIDENCE']) if item.get('AGE_GENDER',False) else 'unexplained'
            new_df.iloc[i,new_df.columns.get_loc('CHAR_'+char+'_AGE_GENDER_PREDICTION_LABEL')] = (item['AGE_GENDER']['PREDICTION']['LABEL']) if item.get('AGE_GENDER',False) else 'unexplained'

            
            
    print('Char_dict_to_columns for Character-'+str(char_number)+' - Done.')
            
            
    return new_df
```

```python
def get_final_char_df(final_char,char_df):
    final_char =final_char.join(char_df.drop([0,1,2,3,4,'COLOR_PALETTE', 'COLOR_LUMINANCE', 'COLOR_TEMPERATURE', 'LOCATION',
       'SHOT_ANGLE', 'SPEECH'],axis=1),how ='outer',lsuffix = 'l', rsuffix = 'r')
    
    
    print('Merge Complete')
    
    return final_char
```

```python
def get_char_age_gender_from_json2(d):
       
    y = get_character_from_json2(d)
    
    list_char = y.drop(['COLOR_PALETTE','COLOR_LUMINANCE','COLOR_TEMPERATURE','LOCATION','SHOT_ANGLE','SPEECH'],axis=1).columns
    
    final_list_l =[]

    final_char = pd.DataFrame()
    for char_number in list_char:
        final_char = get_final_char_df2(final_char,dict_to_columns(y,char_number),list_char)
        

    

        l =[

            'CHAR_'+str(char_number)+'_YMIN',
        'CHAR_'+str(char_number)+'_YMAX',
        'CHAR_'+str(char_number)+'_XMIN',
        'CHAR_'+str(char_number)+'_XMAX',
        'CHAR_'+str(char_number)+'_CHARACTER_PREDICTION_CONFIDENCE',
        'CHAR_'+str(char_number)+'_CHARACTER_PREDICTION_LABEL',
        'CHAR_'+str(char_number)+'_DLIB_CONFIDENCE',
        'CHAR_'+str(char_number)+'_VIDEO_ID',
        'CHAR_'+str(char_number)+'_MINUTE',
        'CHAR_'+str(char_number)+'_SHOT_ID',
        'CHAR_'+str(char_number)+'_SCENE_STARTTIME',
        'CHAR_'+str(char_number)+'_SCENE_ENDTIME',
        'CHAR_'+str(char_number)+'_VIDEO_STARTTIME',
        'CHAR_'+str(char_number)+'_VIDEO_ENDTIME',
        'CHAR_'+str(char_number)+'_QUALITY_PARAMETER_WIDTH',
        'CHAR_'+str(char_number)+'_QUALITY_PARAMETER_HEIGHT',
        'CHAR_'+str(char_number)+'_QUALITY_PARAMETER_PIXEL_COUNT',
        'CHAR_'+str(char_number)+'_AGE_GENDER_PREDICTION_CONFIDENCE',
        'CHAR_'+str(char_number)+'_AGE_GENDER_PREDICTION_LABEL',

        ]
        
        final_list_l.extend(l)
        
        
        print('Char_age_gender_from_json  - Done.')
        
                                   
    return final_char[final_list_l]
    
```

```python
def merge_all_df(l):
    final = pd.DataFrame()
    for df in l:
        final = pd.concat([final,df],axis = 1)
        
        
        
    print('Final Dataframe merge complete.')
    
    return final
```

```python
def get_json2df_collated(path):
    
    df = pd.read_json(path,)
    d = dict(df)
    
    df_char_gender = get_char_age_gender_from_json2(d)
    
    df_shot = get_dict_to_csv_shot(d)

    df_video = get_dict_to_csv_video_id(d)
    
    df_shot_angle = get_dict_to_csv(d,"SHOT_ANGLE")
    
    df_location = get_dict_to_csv(d, 'LOCATION')
    
    df_speech  = get_dict_to_csv(d, 'SPEECH')
    
    
    df_high = get_dict_to_csv_high(d)
    
    df_video_endtime = get_dict_to_csv_video_endtime(d)

    df_color = get_dict_to_csv_color(d)

    q = merge_all_df([df_video,df_shot_angle,df_shot,df_char_gender,df_high,df_speech,df_video,df_color,df_location,df_video_endtime])

    return q
    
```

```python
def get_final_char_df2(final_char,char_df,list_char):
    
    char_df.drop(list(list_char),axis=1)
    
    final_char =final_char.join(char_df.drop(['COLOR_PALETTE', 'COLOR_LUMINANCE', 'COLOR_TEMPERATURE', 'LOCATION',
       'SHOT_ANGLE', 'SPEECH'],axis=1),how ='outer',lsuffix = 'l', rsuffix = 'r')
    return final_char
```

```python
def get_json2csv_collated(path):
    csv_path = path[:-4] + 'csv'
    df = get_json2df_collated(path)
    df.to_csv(csv_path)
```
